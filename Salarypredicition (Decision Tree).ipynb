```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\Nitish\Desktop\Python\salaries.csv")
```

```python
X = data.iloc[:,[0,1,2]]
Y = data.iloc[:,3:]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
n_company = LabelEncoder()
n_job = LabelEncoder()
n_degree = LabelEncoder()
```

```python
X['company_n'] = n_company.fit_transform(X['company'])
X['job_n'] = n_job.fit_transform(X['job'])
X['degree_n'] = n_degree.fit_transform(X['degree'])
```

```python
X = X.drop(['company','job','degree'],axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
dtc.predict([[2,2,0]])
```
